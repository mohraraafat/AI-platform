```python
import torch
```

```python
x = torch.tensor([[1.0, 2.0]], requires_grad=True)
print("Input x:\n", x)
```

```python
w00 = torch.randn((2, 3), requires_grad=True)
b00 = torch.randn((1, 3), requires_grad=True)
z1 = x @ w00 + b00
a1 = torch.relu(z1)
print("\nLayer 1:")
print("z1 =", z1)
print("a1 (ReLU) =", a1)
```

```python
w10 = torch.randn((3, 2), requires_grad=True)
b10 = torch.randn((1, 2), requires_grad=True)
z2 = a1 @ w10 + b10
a2 = torch.sigmoid(z2)
print("\nLayer 2:")
print("z2 =", z2)
print("a2 (Sigmoid) =", a2)
```

```python
combined = a1[:, :2] + a2
a3 = torch.tanh(combined)
print("\nCombine outputs:")
print("Combined sum =", combined)
print("a3 (Tanh) =", a3)
```

```python
w20 = torch.randn((2, 1), requires_grad=True)
b20 = torch.randn((1, 1), requires_grad=True)
z3 = a3 @ w20 + b20
output = z3  # Linear output (no activation)
print("\nOutput Layer:")
print("z3 =", z3)
print("Final output =", output)
```

```python
output.backward(torch.ones_like(output))
print("\n===== Gradients =====")
print("∂Output/∂x =", x.grad)
print("∂Output/∂w00 =", w00.grad)
print("∂Output/∂b00 =", b00.grad)
print("∂Output/∂w10 =", w10.grad)
print("∂Output/∂b10 =", b10.grad)
print("∂Output/∂w20 =", w20.grad)
print("∂Output/∂b20 =", b20.grad)
```
